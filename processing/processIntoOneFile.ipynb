```python
import csv
from bs4 import BeautifulSoup
```

```python


import pandas as pd

df1 = pd.read_csv("../data/wuzzuf_jobs.csv", encoding="utf-8")
df2 = pd.read_csv("../data/indeed_listings.csv", encoding="utf-8")
df3 = pd.read_csv("../data/Linkedin_listings.csv", encoding="utf-8")


df2.head()
```

```python
newdf = pd.DataFrame(
    {
        "url": pd.concat([ df1["job_url"],df2["job_url"],df3["job_url"] ], ignore_index=True),
        "title" :  pd.concat([ df1["title"],df2["title"],df3["job_title"] ], ignore_index=True),
        "company" :  pd.concat([ df1["company_name"],df2["company"],df3["company_name"] ], ignore_index=True),
        "job_type": pd.concat([ df1["job_type"] , pd.Series([None] * len(df2)) , df3["seniority_level"] ], ignore_index=True),
        "employment_type":  pd.concat([ pd.Series([None] * len(df1)), pd.Series([None] * len(df2)), df3["employment_type"] ], ignore_index=True),
        "description" :  pd.concat([ df1["description"],df2["description"],df3["description"] ], ignore_index=True),
        "requirements":  pd.concat([ df1["requirements"] , pd.Series([None] * len(df2)) ,  pd.Series([None] * len(df3)) ] , ignore_index=True),
    }
)
```

```python
newdf.to_csv("merged_jobs.csv", index=False, encoding="utf-8")
```

```python

# simple cleaning
def clean_html(text):
    return BeautifulSoup(str(text), "html.parser").get_text(" ", strip=True)

newdf["description"] = newdf["description"].apply(clean_html)
newdf["requirements"] = newdf["requirements"].apply(clean_html)

newdf = newdf.fillna('') 

text_cols = ['title', 'company', 'description', 'requirements']
for col in text_cols:
    newdf[col] = newdf[col].str.strip()


```

```python

newdf.to_csv("merged_jobs_clean.csv", index=False, encoding="utf-8-sig") 
```
